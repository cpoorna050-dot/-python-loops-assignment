```python
print("===== Task 1: Find Maximum and Minimum =====")
def temperature_MAX_MIN():
    temperatures = [28, 32, 35, 29, 31, 27, 30]

    if not temperatures:
        print("List is empty")
        return

    # Start with the first value in the list
    Highest = temperatures[0]
    Lowest = temperatures[0]
    Days = 0
    for temp in temperatures:
        if temp > Highest:
            Highest = temp
        if temp < Lowest:
            Lowest = temp
        if temp > 30:
            Days += 1

    print(F"Highest temperature:{Highest}°C")
    print(F"Lowest temperature:{Lowest}°C")
    print(f"Hot days(>30°C):{Days}")
temperature_MAX_MIN()

print("===== Task 2: Count Hot Days =====")
def Hot_Days():
    temperatures = [28, 32, 35, 29, 31, 27, 30]

    if not temperatures:
        print("List is empty")
        return
    Days = 0
    for temp in temperatures:
        if temp > 30:
            Days += 1

    print(f"Hot days(>30°C):{Days}")
Hot_Days()

print("===== Task 3: Alert System =====")
def Alert_system():
    temperatures = [28, 32, 35, 40, 31, 33, 30]
    if not temperatures:
        print("List is empty")
        return
    Days = 0
    pos = 0
    for i,temp in enumerate(temperatures):
        if temp >= 40:
            Days +=1
            pos = i + 1
            break

    print(f"Alter extreme temperature {temp}°C deteced on day {pos}")
Alert_system()
```
